separating Numeric and Non-Numeric Features

```python
numeric_data=df.select_dtypes(include=['float64', 'int64'])
non_numeric=df.select_dtypes(include = ['object'])

numeric_data1=dv.select_dtypes(include=['float64', 'int64'])
non_numeric1=dv.select_dtypes(include = ['object'])
```

```python
non_numeric.columns
```

```python
def convert_plus_format(value):
    if isinstance(value, str) and ('+' in value or '-' in value): # Fix the condition
        if '+' in value:
            parts = value.split('+')
        else:
            parts = value.split('-')
        return int(parts[0]) + int(parts[1])
    return value
```

```python
for col in non_numeric:
    non_numeric[col] = non_numeric[col].apply(convert_plus_format)
```

Training and evaluating models

```python
#from sklearn.ensemble import HistGradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
Xtrain_imputed = imputer.fit_transform(Xtrain)
Xtest_imputed = imputer.transform(Xtest)
```

```python
sc=StandardScaler()
Xtrain=sc.fit_transform(Xtrain_imputed)
Xtest=sc.transform(Xtest_imputed)
```

```python
models = {#'hgb':HistGradientBoostingRegressor().fit(Xtrain, Ytrain),
         'rf':RandomForestRegressor(max_depth=7),
         'xgb':XGBRegressor(),
'gb':GradientBoostingRegressor(n_estimators=100)
}
```

Evaluation

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import pickle as pkl
```

```python
for name, model in models.items(): # Iterate over key-value pairs
    scores = cross_val_score(model, Xtrain, Ytrain, cv=5, scoring='neg_mean_absolute_error')
    print(f'{name} Cross-Validation MAE: {-scores.mean()}')
    model.fit(Xtrain, Ytrain)
    pkl.dump(model, open(f'/content/{model.__class__.__name__}.pkl', 'wb'))

    y_pred = model.predict(Xtest)
```

Grid Search with Cross Validation

```python
from sklearn.model_selection import GridSearchCV, KFold
```

```python
cv=KFold(n_splits=3)
gb = GradientBoostingRegressor()
PARAMETERS_gb ={
"max_depth":[2,5, 6, 12],
"min_samples_leaf":[1,5,10],
"min_samples_split":[2,5,10],
# "min_child_weight":[1,5,15],​
"learning_rate":[0.3, 0.1, 0.03],
"n_estimators":[100,500,1000]}
model_gs=GridSearchCV(gb,param_grid=PARAMETERS_gb,cv=cv,scoring="accuracy")
model_gs.fit(Xtrain, Ytrain)
pkl.dump(model_gs, open('/content/' + model_gs.__class__.__name__ + '.pkl', 'wb'))
y_pred = model_gs.predict(Xtest)
print(model_gs.__class__.__name__, mean_absolute_error(Ytest, y_pred))
```
